# Q1 Exploratory Data Analysis

```python
# Importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# read and describe the data set

cd=pd.read_csv('clean_kaggle_data.csv')
cd.describe()
```

```python
# printing the unique values of age, coding experience and machine learning and saving them in separate lists

age_list=cd['Q1'].unique()
code_exp=cd['Q15'].unique()
ml_exp=cd['Q23'].unique()
print('The unique age values are') 
print(age_list)
print('The unique Coding Experience values are') 
print(code_exp)
print('The unique Machine Learning Experience values are')
print(ml_exp)
```

```python
#sorting the unique values of age

age_list.sort()
```

```python
# assigning an average value to each experince inplace of interval in 2 new columns and decsribe the new data set

c_e=[1.5,0,0.5,25,4,7.5,15,0]
m_e=[1.5,0,2.5,0.5,12.5,3.5,4.5,7.5,25]

cd["code_exp"]=0
cd["ml_exp"]=0

for a in range(len(code_exp)):
    cd['code_exp'][cd['Q15']==code_exp[a]]=c_e[a]
    
for b in range(len(ml_exp)):
    cd['ml_exp'][cd['Q23']==ml_exp[b]]=m_e[b]
    
cd.describe()
```

```python
# calculating the average coding experience for each group (Male, female and others)

avg_exp_code_m=[]
avg_exp_code_f=[]
avg_exp_code_o=[]
for i in range(len(age_list)):
    avg_exp_code_m.append(cd[(cd['Q2']=="Male") & (cd['Q1']==age_list[i])]['code_exp'].mean())
    avg_exp_code_f.append(cd[(cd['Q2']=="Female") & (cd['Q1']==age_list[i])]['code_exp'].mean())
    avg_exp_code_o.append(cd[(cd['Q2']!="Male") & (cd['Q2']!="Female") & (cd['Q1']==age_list[i])]['code_exp'].mean())
```

```python
# calculating the average Machine Learning experience for each group (Male, female and others)

avg_exp_ml_m=[]
avg_exp_ml_f=[]
avg_exp_ml_o=[]
for i in range(len(age_list)):
    avg_exp_ml_m.append(cd[(cd['Q2']=="Male") & (cd['Q1']==age_list[i])]['ml_exp'].mean())
    avg_exp_ml_f.append(cd[(cd['Q2']=="Female") & (cd['Q1']==age_list[i])]['ml_exp'].mean())
    avg_exp_ml_o.append(cd[(cd['Q2']!="Male") & (cd['Q2']!="Female") & (cd['Q1']==age_list[i])]['ml_exp'].mean())
```

```python
# plotiing the average coding and machine learning experience for each group (Male, female and others)

x = np.arange(len(age_list))
width = 0.7
fig, ax = plt.subplots(figsize=(13, 8))


ax.bar(x - width/3, avg_exp_code_m, width/3, label='Coding-Male')
ax.bar(x , avg_exp_code_f, width/3, label='Coding-Female')
ax.bar(x + width/3, avg_exp_code_o, width/3, label='Coding-Other')

ax.bar(x - width/3, avg_exp_ml_m, width/3, label='ML-Male')
ax.bar(x , avg_exp_ml_f, width/3, label='ML-Female')
ax.bar(x + width/3, avg_exp_ml_o, width/3, label='ML-Other')

ax.set_ylabel('Average Experience')
ax.set_xlabel('Age Groups')
ax.set_title('Average Coding and ML Experience by Age group and Gender')
ax.set_xticks(x)
ax.set_xticklabels(age_list)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
ax.legend()
plt.tight_layout()
plt.show()
```

#### Note:- In this bar chart all the bars are starting from 0 y-value (the bars are of overlapping nature)

#### Here we can observe that the Machine Learning experience follows the same pattern as coding experience in terms of gender
#### The other conclusion which can be drawn is that the coding experience increases with age with not the ML experience, it is almost not unaffected by the age
#### Also it can be observed that the average experience of Males is much higher than Females in almost every age group

```python
# box plot distributions of salary based on Geographical Locations (countries)

fig, ax = plt.subplots(figsize=(13, 10))
sns.boxplot(x=cd['Q3'],y=cd['Q10'],data=cd,width=0.5)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
ax.set_xlabel("Countries")
ax.set_ylabel("Salary Distribution")
ax.set_title("Dsitribution of Salary based on Geographical Location")
plt.tight_layout()
plt.show()
```

#### Here we can observe that for some countries like USA, Australia, Israel, Switzerland and Norway the median salary is pretty high which as also reasonable as these are all developed countries.
#### While on the other hand, for countries like Pakistan, Nigeria, Indonesia, Egypt, Vietnam, Iran and Algeria the salaries are pretty low and the distribution is also squeezed

```python
# box plot distributions of salary based on Education

fig, ax = plt.subplots(figsize=(13, 8))
sns.boxplot(x=cd['Q4'],y=cd['Q10'],data=cd,width=0.5)
ax.set_xticklabels(ax.get_xticklabels(), rotation=20, ha="right")
ax.set_xlabel("Education")
ax.set_ylabel("Salary Distribution")
ax.set_title("Dsitribution of Salary based on Education")
plt.tight_layout()
plt.show()
```

# Q2 T-Test

```python
# Importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy.stats as st
from scipy.stats import norm
```

```python
# read the data set

cd=pd.read_csv('clean_kaggle_data.csv')
```

```python
# segregating the dataset into three groups based on gender

male=cd[cd['Q2']=="Male"]
female=cd[cd['Q2']=="Female"]
```

### Q2(a) decriptve analysis of each group

```python
print(male['Q10'].describe())
```

```python
print(female['Q10'].describe())
```

```python
# box plot distribution of each group

MF=pd.concat([male,female])
sns.boxplot(x=MF['Q2'],y=MF['Q10'],data=MF,width=0.5)
plt.tight_layout()
plt.xlabel("Gender")
plt.ylabel("Salary")
plt.title("Dsitribution of Salary based on Gender")
plt.show()
```

#### the salary distribution is almost similar for both the genders

#### there are some outliers as well in both the groups

 ### Q2(b) Two-Sample T-Test with 0.05 threshold

#### Assumptions for conducting the T-test
##### Data values must be independent (Measurements for one observation do not affect measurements for any other observation.)
##### Data in each group must be obtained via a random sample from the population.
##### Data in each group is normally distributed. (which is not in our case)
##### Data values are continuous.
##### The variances for the two independent groups are equal. (almost equal in our case)

#### We cannot perform the T-Test as the data for the two groups is not normally distributed (visible in the box plot distributions)
###### If we still do the T-Test, the results will be like this

```python
# Two-sample T-Test

t,p=st.ttest_ind(male['Q10'],female['Q10'],equal_var=False)
print('t value is '+ str(t))
print('p value is '+ str(p))
```

#### here we are assuming that the mean salaries of the two groups are equal, which is our null hypothesis

```python
# checking whether the null hypothesis is valid or not 

if p<0.05:
    print('Null Hypothesis is rejected')
else:
    print('Null Hypothesis is accepted')
```

#### since the p value is less than the threshold value, so the null hypothesis is rejected

#### which means that there is a significant difference in mean salaries of the two groups

## Q2(c) Bootstrapping

```python
# Bootstrapping the data, calculating the mean salary for Males and replicating the process 1000 times

np.random.seed(123)
rnd_samp_m=[]
for i in range(1000):
    rnd_samp_m.append(np.random.choice(male['Q10'],size=int(((male['Q10'].count()).round()))).mean())
rnd_samp_m=np.array(rnd_samp_m)
print('The mean salary from the Bootstrapped data for male is ' +str(round(rnd_samp_m.mean(),2)))
print('The standard deviation of salary from the Bootstrapped data for male is ' +str(round(rnd_samp_m.std(),2)))
```

### Bootstrapped Distribution for Salary of Males

```python
# histogram along with the Gaussian curve for salary distribution of Males

n,bins,b=plt.hist(rnd_samp_m,bins=15,density=1)
best_fit1=norm.pdf(bins, rnd_samp_m.mean(), rnd_samp_m.std())
plt.plot(bins, best_fit1,color='r',marker='.')
plt.tight_layout()
plt.xlabel("Mean Salary")
plt.ylabel("Frequency")
plt.title("Dsitribution of Mean Salary for 1000 boot strapped samples for Males")
plt.show()
```

```python
# Calculating the confidence interval

confidence_level = 0.95
degrees_freedom = len(rnd_samp_m) - 1
sample_mean = np.mean(rnd_samp_m)
sample_standard_error = st.sem(rnd_samp_m)
confidence_interval = st.t.interval(confidence_level, degrees_freedom, sample_mean, sample_standard_error)
print('The mean salary of Males will lie between '+ str(round(confidence_interval[0],2))+ ' and '+ str(round(confidence_interval[1],2))+' with a confidence of 95%')
```

```python
# Bootstrapping the data, calculating the mean salary for Females and replicating the process 1000 times

np.random.seed(123)
rnd_samp_f=[]
for i in range(1000):
    rnd_samp_f.append(np.random.choice(female['Q10'],size=int(((female['Q10'].count()).round()))).mean())
rnd_samp_f=np.array(rnd_samp_f)
print('The mean salary from the Bootstrapped data for male is ' +str(round(rnd_samp_f.mean(),2)))
print('The standard deviation of salary from the Bootstrapped data for male is ' +str(round(rnd_samp_f.std(),2)))
```

### Bootstrapped Distribution for Salary of Females

```python
# histogram along with the Gaussian curve for salary distribution of Females

n,bins,b=plt.hist(rnd_samp_f,bins=15,density=1)
best_fit2=norm.pdf(bins, rnd_samp_f.mean(), rnd_samp_f.std())
plt.plot(bins, best_fit2,color='r',marker='.')
plt.tight_layout()
plt.xlabel("Mean Salary")
plt.ylabel("Frequency")
plt.title("Dsitribution of Mean Salary for 1000 boot strapped samples for Females")
plt.show()
```

```python
# Calculating the confidence interval

confidence_level = 0.95
degrees_freedom = len(rnd_samp_f) - 1
sample_mean = np.mean(rnd_samp_f)
sample_standard_error = st.sem(rnd_samp_f)
confidence_interval = st.t.interval(confidence_level, degrees_freedom, sample_mean, sample_standard_error)
print('The mean salary of Females will lie between '+ str(round(confidence_interval[0],2))+ ' and '+ str(round(confidence_interval[1],2))+' with a confidence of 95%')
```

#### Both of the bootstrapped distributions are normally distributed

```python
# calculating the difference in means

difference=rnd_samp_m-rnd_samp_f
```

### Distribution for Difference in means of salaries of Males and Females

```python
# histogram along with the Gaussian curve for difference in means of salaries of Males and Females

n,bins,b=plt.hist(difference,bins=15,density=1)
best_fit3=norm.pdf(bins, difference.mean(), difference.std())
plt.plot(bins, best_fit3,color='r',marker='.')
plt.tight_layout()
plt.xlabel("Mean Salary Difference")
plt.ylabel("Frequency")
plt.title("Distribution for Difference in means of salaries of Males and Females")
plt.show()
```

#### As observed the distribution is normally distributed and it should be, as it is the difference of the normal distributions only

### Q2(d) Two-Sample T-Test with 0.05 threshold on bootstrapped data of the two groups  

```python
# Two-sample T-Test

t_b,p_b=st.ttest_ind(rnd_samp_m,rnd_samp_f,equal_var=False)
print('t value is '+ str(t_b))
print('p value is '+ str(p_b))
```

#### here we are assuming that the mean salaries of the two groups are equal, which is our null hypothesis

```python
# checking whether the null hypothesis is valid or not 

if p_b<0.05:
    print('Null Hypothesis is rejected')
else:
    print('Null Hypothesis is accepted')
```

#### since the p value is less than the threshold value, so the null hypothesis is rejected

#### which means that there is a significant difference in mean salaries of the two groups

# Q3 ANOVA

```python
# Importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy.stats as st
from scipy.stats import norm
```

```python
# read the data set

cd=pd.read_csv('clean_kaggle_data.csv')
```

```python
# segregating the dataset into three groups based on highest level of formal education

B=cd[cd['Q4']=="Bachelorâ€™s degree"]
M=cd[cd['Q4']=="Masterâ€™s degree"]
D=cd[cd['Q4']=="Doctoral degree"]
```

### Q3(a) decriptve analysis of each group

```python
# decription of each group

print(B['Q10'].describe())
```

```python
print(M['Q10'].describe())
```

```python
print(D['Q10'].describe())
```

#### the mean salary increases with increasing level of education

```python
# box plot distribution of each group

BMD=pd.concat([B,M,D])
sns.boxplot(x=BMD['Q4'],y=BMD['Q10'],data=BMD,width=0.5)
plt.tight_layout()
plt.xlabel("Highest Level of Formal Education")
plt.ylabel("Salary")
plt.title("Dsitribution of Salary based on Formal Education")
plt.show()
```

#### the salary is more evenly distributed with higher level of education

#### there are some outliers as well in all the three groups

 ### Q3(b) One way ANOVA on three groups

#### Assumptions for conducting the ANOVA
##### Each group sample is drawn from a normally distributed population
##### All populations have a common variance
##### All samples are drawn independently of each other
##### Within each sample, the observations are sampled randomly and independently of each other
##### Factor effects are additive

```python
# One way ANOVA on three groups

F,p=st.f_oneway(B['Q10'],M['Q10'],D['Q10'])
print('F value is '+ str(F))
print('p value is '+ str(p))
```

#### As observed the F value is pretty high and also the p value is drastically small. So we can conclude that there is a huge variation between the three groups.
